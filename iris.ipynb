```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load the iris dataset directly from seaborn
df = sns.load_dataset("iris")

# display first few rows
df.head()
```

```python
print("Shape of dataset:", df.shape)
print("\nInfo:")
print(df.info())

print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Drop rows with missing values (if any)
df_clean = df.dropna().reset_index(drop=True)
print("After cleaning:", df_clean.shape)
```

```python
df_clean.describe()
```

```python
group_means = df_clean.groupby('species').mean()
print("Mean values by species:\n", group_means)
```

```python
group_means['petal_length'].plot(kind='bar', color=['skyblue','orange','lightgreen'])
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.hist(df_clean['sepal_length'], bins=15, edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df_clean, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```

```python
plt.plot(df_clean.index, df_clean['sepal_length'], color='purple')
plt.title('Simulated Trend of Sepal Length')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
df_clean.to_csv("iris_cleaned.csv", index=False)
print("Saved iris_cleaned.csv successfully.")
```
